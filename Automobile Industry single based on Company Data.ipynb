```python
import pandas as pd
import numpy as np
import random

manufacturing_units=['Becharaji, Mehsana, Gujarat.','Gurugram and Manesar, Gurgaon, Haryana.']
models=[ 'Wagon R-1999','Alto-2000','Swift-2005','Ertiga-2012','Celerio-2014',
        'Ciaz-2014','Baleno-2015','S-Cross-2015','Vitara Brezza-2016','Dzire-2017',
        'Ignis-2017','XL6-2019','S-Presso-2019']
makes=['Maruti Suzuki']
types=['Petrol','Diesel']


columns=['YEAR','VEHICLE ID','MANUFACTURING UNIT','MODEL','MAKE',
         'TYPE','SALE','REVENUE,Rs (in Crores)','EBIT']

df=pd.DataFrame(columns=columns)


for i in range(500):
    year=random.randint(2005,2021)
    manufacturing_unit=random.choice(manufacturing_units)
    model=random.choice(models)
    make=random.choice(makes)
    type=random.choice(types)
    sale=random.randint(10000,99999)
    revenue=random.uniform(3122.90,79031.40)
    Earnings_Before_Interest_and_Taxes=random.uniform(-2643.42,7197.70)
    df.loc[i]=[year,i,manufacturing_unit,model,make,type,sale,revenue,Earnings_Before_Interest_and_Taxes]
    df.to_csv('test1.csv')
    
    
```
